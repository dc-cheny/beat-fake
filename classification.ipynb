```python
import os
import pandas as pd
import numpy as np
from imblearn.over_sampling import SMOTE
from imblearn.combine import SMOTEENN
import sklearn
```

### 数据读取及预处理

```python
fdata = pd.read_csv("./all_parts_features.csv",sep=',')
```

```python
fdata
```

```python
fdata = fdata.drop(['rect_left'],axis=1)
```

```python
fdata
```

#### SVM分类处理

- 样本平衡处理

```python
from sklearn import svm
```

```python
#划分训练集和测试集并对训练集进行SMOTE_ENN平衡处理
train_X,test_X,train_y,test_y =sklearn.model_selection.train_test_split(np.array(fdata)[:,1:26], np.array(fdata)[:,26], random_state=15, train_size=0.6,test_size=0.4)
```

```python
train_y = train_y.astype(np.int32)
```

```python
smo_enn = SMOTEENN(random_state=15)
X_smo, y_smo = smo_enn.fit_sample(train_X, train_y)
```

```python
print("原数据正负样本量：",train_y[train_y==0].size,":",train_y[train_y==1].size)
print("SMOTEENN处理后正负样本量：",y_smo[y_smo==0].size,":",y_smo[y_smo==1].size)
```

- SVM模型

```python
#SVM选用线性核函数
classifier=svm.SVC(C=3,kernel='linear',gamma=3,decision_function_shape='ovr')
classifier.fit(X_smo, y_smo)
```

```python
print("训练集：",classifier.score(X_smo, y_smo))
print("测试集：",classifier.score(test_X,test_y))
```
